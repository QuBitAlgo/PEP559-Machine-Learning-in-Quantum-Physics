```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Extract first 100 rows (binary classification)
sub_X = iris_df.iloc[0:100, [0, 2]].values  # Selecting two features
sub_y = iris_df.iloc[0:100, 4].values  # Class labels

# Correct the label transformation
sub_y = np.where(sub_y == 'Iris-setosa', 0, 1)

# Define perceptron functions
def initialize_weights(n_features, random_state=1):
    rgen = np.random.RandomState(random_state)
    weights = rgen.normal(loc=0.0, scale=0.01, size=n_features)
    bias = np.float64(0.)
    return weights, bias

def net_input(X, weights, bias):
    return np.dot(X, weights) + bias

def predict(X, weights, bias):
    return np.where(net_input(X, weights, bias) >= 0.0, 1, 0)

def train_perceptron(X, y, eta=0.01, n_iter=50, random_state=1):
    weights, bias = initialize_weights(X.shape[1], random_state)
    errors_list = []
    
    for _ in range(n_iter):
        errors = 0
        for xi, target in zip(X, y):
            update = eta * (target - predict(xi, weights, bias))
            weights += update * xi
            bias += update
            errors += int(update != 0.0)
        errors_list.append(errors)
    
    return weights, bias, errors_list

# Train perceptron model
weights, bias, errors = train_perceptron(sub_X, sub_y, eta=0.1, n_iter=10)

# Plot number of updates per epoch
plt.plot(range(1, len(errors) + 1), errors, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of updates')
plt.title('Perceptron Learning')
plt.show()
```

# Apply the Adaline learning to classify the same dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Extract first 100 rows (binary classification)
sub_X = iris_df.iloc[0:100, [0, 2]].values  # Selecting two features
sub_y = iris_df.iloc[0:100, 4].values  # Class labels

# Correct the label transformation
sub_y = np.where(sub_y == 'Iris-setosa', -1, 1)  # Traditionally, adaline uses -1 and 1 labels

def initialize_weights(n_features, random_state=1):
    rgen = np.random.RandomState(random_state)
    weights = rgen.normal(loc=0.0, scale=0.01, size=n_features)
    bias = np.float64(0.)
    return weights, bias

def net_input(X, weights, bias):
    return np.dot(X, weights) + bias

def activation(X):
    return X  # Linear activation function for Adaline

def compute_loss(y, output):
    return ((y - output) ** 2).sum() / 2  # Mean Squared Error (MSE) # sometimes people introduce 1/2 for convenience

def train_adaline(X, y, eta=0.01, n_iter=50, random_state=1):
    weights, bias = initialize_weights(X.shape[1], random_state)
    losses = []
    
    for _ in range(n_iter):
        net_out = net_input(X, weights, bias)
        output = activation(net_out)
        errors = y - output
        weights += eta * X.T.dot(errors) / X.shape[0]  # Update using batch gradient descent
        bias += eta * errors.mean()  # Bias update
        losses.append(compute_loss(y, output))  # Track loss
    
    return weights, bias, losses

# Train Adaline model
weights, bias, losses = train_adaline(sub_X, sub_y, eta=0.02, n_iter=100)

# Plot the loss over epochs
plt.plot(range(1, len(losses) + 1), losses, marker='.')
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error')
plt.title('Adaline Learning - Loss Reduction')
plt.show()
```

# Adaptive learning rate

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Extract first 100 rows (binary classification)
sub_X = iris_df.iloc[0:100, [0, 2]].values  # Selecting two features
sub_y = iris_df.iloc[0:100, 4].values  # Class labels

# Correct the label transformation
sub_y = np.where(sub_y == 'Iris-setosa', -1, 1)  # Adaline uses -1 and 1 labels

def initialize_weights(n_features, random_state=1):
    rgen = np.random.RandomState(random_state)
    weights = rgen.normal(loc=0.0, scale=0.01, size=n_features)
    bias = np.float64(0.)
    return weights, bias

def net_input(X, weights, bias):
    return np.dot(X, weights) + bias

def activation(X):
    return X  # Linear activation function for Adaline

def compute_loss(y, output):
    return ((y - output) ** 2).sum() / 2  # Mean Squared Error (MSE)

def train_adaline_adaptive(X, y, c1=0.01, c2=1, n_iter=50, random_state=1):
    weights, bias = initialize_weights(X.shape[1], random_state)
    losses = []
    
    for i in range(n_iter):
        eta = c1 / (i + c2)  # Adaptive learning rate
        net_out = net_input(X, weights, bias)
        output = activation(net_out)
        errors = y - output
        weights += eta * X.T.dot(errors) / X.shape[0]  # Update using batch gradient descent
        bias += eta * errors.mean()  # Bias update
        losses.append(compute_loss(y, output))  # Track loss
    
    return weights, bias, losses

# Train Adaline model with adaptive learning rate
weights, bias, losses = train_adaline_adaptive(sub_X, sub_y, c1=0.15, c2=1, n_iter=100)

# Plot the loss over epochs
plt.plot(range(1, len(losses) + 1), losses, marker='.')
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error')
plt.title('Adaline Learning with Adaptive Learning Rate')
plt.show()
```
